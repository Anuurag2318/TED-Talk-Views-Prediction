- To improve the prediction model, we can consider including features such as speaker ratings on TED Talks. 

- These ratings can help in understanding the quality of the talk and the popularity of the speaker, which could be important factors in predicting the number of views.

```python
# Create a new column 'video_rating' in the main DataFrame and assign the categories

tedtalks['video_rating'] = ""
tedtalks.loc[tedtalks['comments'] <= 50, 'video_rating'] = 1
tedtalks.loc[(tedtalks['comments'] > 50) & (tedtalks['comments'] <= 120), 'video_rating'] = 2
tedtalks.loc[(tedtalks['comments'] > 120) & (tedtalks['comments'] <= 200), 'video_rating'] = 3
tedtalks.loc[(tedtalks['comments'] > 200) & (tedtalks['comments'] <= 300), 'video_rating'] = 4
tedtalks.loc[tedtalks['comments'] > 300, 'video_rating'] = 5

# check the dataset
tedtalks.sample(2)
```

Comments can be a measure of engagement and audience interest in the content, as people who leave comments are often invested in the topic and want to express their thoughts or feedback.

- In the case of TED Talks, comments can be an indicator of the quality and impact of the speaker's presentation, as well as the relevance and resonance of the topic with the audience.

- Therefore, it can be hypothesized that videos with higher comments are more likely to be well-received and impactful, and thus may be rated more highly by viewers.

# **model implementation and pre-processing**

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()

tedtalks['event']=encoder.fit_transform(tedtalks['event'])
tedtalks['main_speaker']=encoder.fit_transform(tedtalks['main_speaker'])
tedtalks['name']=encoder.fit_transform(tedtalks['name'])
tedtalks['title']=encoder.fit_transform(tedtalks['title'])
tedtalks['speaker_popularity']=encoder.fit_transform(tedtalks['speaker_popularity'])
tedtalks['tags']=encoder.fit_transform(tedtalks['tags'])
tedtalks['ratings']=encoder.fit_transform(tedtalks['ratings'])
```

```python
tedtalks.info()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

tedtalks['views'] = np.log1p(tedtalks['views'])  # Use log1p to avoid log(0) issues

# Redefine the target variable
y = tedtalks['views']


# split the Dataset into independent(x) and dependent(y) Dataset

X = tedtalks.drop(columns=['views','duration','published_date','film_date'])
y = tedtalks['views']
# Split the data again
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=23)

# split sizes
print(xtrain.shape)
print(xtest.shape)
```

# **LinearRegression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Initialize the model
lr= LinearRegression()

# Fit the model
lr.fit(xtrain, ytrain)

# Predict
ytrain_pred = lr.predict(xtrain)
ytest_pred = lr.predict(xtest)

# Calculate metrics for training data
r2_train = r2_score(ytrain, ytrain_pred)
adj_r2_train = 1 - (1 - r2_train) * ((len(ytrain) - 1) / (len(ytrain) - xtrain.shape[1] - 1))
mae_train = mean_absolute_error(ytrain, ytrain_pred)
mse_train = mean_squared_error(ytrain, ytrain_pred)
rmse_train = np.sqrt(mse_train)

# Calculate metrics for testing data
r2_test = r2_score(ytest, ytest_pred)
adj_r2_test = 1 - (1 - r2_test) * ((len(ytest) - 1) / (len(ytest) - xtest.shape[1] - 1))
mae_test = mean_absolute_error(ytest, ytest_pred)
mse_test = mean_squared_error(ytest, ytest_pred)
rmse_test = np.sqrt(mse_test)

# Print the performance metrics
print("Training data R2 and Adjusted R2 Score")
print("-----------------------------------------")
print(f"R2 score {r2_train}")
print(f"Adjusted R2 score {adj_r2_train}")
print()

print("Testing data R2 and Adjusted R2 Score")
print("-----------------------------------------")
print(f"R2 score {r2_test}")
print(f"Adjusted R2 score {adj_r2_test}")
print()

print("The performance metrics")
print("-----------------------------------------")
print(f"MAE {mae_test}")
print(f"MSE {mse_test}")
print(f"RMSE {rmse_test}")
```

```python
# Plot actual vs predicted prices
plt.figure(figsize=(10, 6))
plt.plot(ytest.values, label='Actual Value',)
plt.plot(ytest_pred, label='Predicted Value',)
plt.title('Actual vs Predicted Value')
plt.xlabel('Sample Index')
plt.ylabel('Views')
plt.legend()
plt.show()
```

# **RandomForest**

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Initialize the model
rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model
rf.fit(xtrain, ytrain)

# Predict
ytrain_pred = rf.predict(xtrain)
ytest_pred = rf.predict(xtest)

# Calculate metrics for training data
r2_train = r2_score(ytrain, ytrain_pred)
adj_r2_train = 1 - (1 - r2_train) * ((len(ytrain) - 1) / (len(ytrain) - xtrain.shape[1] - 1))
mae_train = mean_absolute_error(ytrain, ytrain_pred)
mse_train = mean_squared_error(ytrain, ytrain_pred)
rmse_train = np.sqrt(mse_train)

# Calculate metrics for testing data
r2_test = r2_score(ytest, ytest_pred)
adj_r2_test = 1 - (1 - r2_test) * ((len(ytest) - 1) / (len(ytest) - xtest.shape[1] - 1))
mae_test = mean_absolute_error(ytest, ytest_pred)
mse_test = mean_squared_error(ytest, ytest_pred)
rmse_test = np.sqrt(mse_test)

# Print the performance metrics
print("Training data R2 and Adjusted R2 Score")
print("-----------------------------------------")
print(f"R2 score {r2_train}")
print(f"Adjusted R2 score {adj_r2_train}")
print()

print("Testing data R2 and Adjusted R2 Score")
print("-----------------------------------------")
print(f"R2 score {r2_test}")
print(f"Adjusted R2 score {adj_r2_test}")
print()

print("The performance metrics")
print("-----------------------------------------")
print(f"MAE {mae_test}")
print(f"MSE {mse_test}")
print(f"RMSE {rmse_test}")
```

```python
# Plot actual vs predicted prices
plt.figure(figsize=(10, 6))
plt.plot(ytest.values, label='Actual Value',)
plt.plot(ytest_pred, label='Predicted Value',)
plt.title('Actual vs Predicted Value')
plt.xlabel('Sample Index')
plt.ylabel('Views')
plt.legend()
plt.show()
```
